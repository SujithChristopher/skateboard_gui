```python
while True:
    if kinectColor.has_new_color_frame():
        color_frame = kinectColor.get_last_color_frame()
        color_frame = np.reshape(color_frame, (1080, 1920, 4))
        color_frame = cv2.flip(color_frame, 1)
        break
```
